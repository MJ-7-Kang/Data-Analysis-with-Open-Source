## 13-2 ì›¹ ìŠ¤í¬ë˜í•‘ í•¨ìˆ˜ ì •ì˜ ë° ì‹¤í–‰

```python
import google_colab_selenium as gs
from lxml import html

## URLë¡œë¶€í„° í˜ì´ì§€ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°
def get_page(driver, url):
    # url í˜ì´ì§€ë¡œ ì´ë™
    driver.get(url)
    # í•´ë‹¹ í˜ì´ì§€ì˜ htmlì„ page_contentë¡œ ì €ì¥
    page_content = driver.page_source
    # page_contentë¥¼ lxmlì˜ html ê°ì²´ë¡œ ë³€í™˜í•˜ê³  treeë¡œ ì €ì¥
    tree = html.fromstring(page_content)
    return tree

## HTML íŠ¸ë¦¬ì—ì„œ ì œëª© ì¶”ì¶œ
def extract_titles(tree):
    ## td-subjectë¥¼ í´ë˜ìŠ¤ë¡œ ê°€ì§€ëŠ” td íƒœê·¸ > aíƒœê·¸ > strong > text() ì„ xpathë¡œ êµ¬í•˜ê³  titlesë¡œ ì €ì¥
    titles= tree.xpath('//td[@class="td-subject"]/a/strong/text()')
    return titles

## ì œëª© ëª©ë¡ ì¶œë ¥
def print_titles(titles):
    for title in titles:
        print(title)

## Chrome ë“œë¼ì´ë²„ ì´ˆê¸°í™”
driver = gs.Chrome()

## ì»´í“¨í„°ê³¼í•™ê³¼ ê²Œì‹œíŒ ì²«ë²ˆì§¸ í˜ì´ì§€ ì œëª© ë°ì´í„° ìˆ˜ì§‘
board_name = 'ì»´í“¨í„°ê³¼í•™ê³¼'
board_url = 'https://cs.knou.ac.kr/cs1/4794/subview.do'
tree = get_page(driver, board_url)
titles = extract_titles(tree)
print_titles(titles)
```

## 13-3 ë‹¤ì¤‘ í˜ì´ì§€ ë°ì´í„° ìˆ˜ì§‘ ë° ë°ì´í„°í”„ë ˆì„ ìƒì„±

```python
#import google_colab_selenium as gs
from lxml import html

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

## ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™í•˜ê³  tree ì •ë³´ ë°˜í™˜
def move_to_next_page(driver):
    # _listNext í´ë˜ìŠ¤ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì—˜ë¦¬ë¨¼íŠ¸ ì„ íƒ
    next_page_link = driver.find_element(By.CLASS_NAME, "_listNext")
    if next_page_link and next_page_link.is_enabled():
        # ë‹¤ìŒ í˜ì´ì§€ ë²„íŠ¼ í´ë¦­
        next_page_link.click()
        # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "board-table"))
        )
        page_content = driver.page_source
        tree = html.fromstring(page_content)
        return tree
    else:
        return None

## ì—¬ëŸ¬ í˜ì´ì§€ì—ì„œ ê²Œì‹œë¬¼ ì œëª© ê°€ì ¸ì˜¤ê¸°
def get_board_titles(url, pages, verbose=False):
    driver = gs.Chrome()
    board_titles = []
    for page in pages:
        if verbose:
          print(f"=== í˜ì´ì§€ {page} ì²˜ë¦¬ì¤‘ì…ë‹ˆë‹¤. ===")
        if page == 1:
            # 13-2 ì—ì„œ ì‘ì„±í•œ get_page í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ í˜„ì¬ í˜ì´ì§€ tree êµ¬í•˜ê¸°
            tree = get_page(driver, url)
        else:
            # ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™í•˜ê³  í•´ë‹¹ í˜ì´ì§€ tree êµ¬í•˜ê¸°
            tree = move_to_next_page(driver)
            if tree is None:
                break
        if tree is not None:
            titles = extract_titles(tree)
            board_titles.extend(titles)
            if verbose:
              print_titles(titles)
    driver.quit()  # Chrome ë“œë¼ì´ë²„ ì¢…ë£Œ
    return board_titles

## ì»´í“¨í„°ê³¼í•™ê³¼ ê²Œì‹œíŒ 1 í˜ì´ì§€ ~ 10 í˜ì´ì§€ì˜ ì œëª© ë°ì´í„° ìˆ˜ì§‘
board_name = 'ì»´í“¨í„°ê³¼í•™ê³¼'
board_url = 'https://cs.knou.ac.kr/cs1/4794/subview.do'
board_titles = get_board_titles(url=board_url, pages=range(1,11), verbose=True)
## ì»´í“¨í„°ê³¼í•™ê³¼ ì œëª© ë°ì´í„°í”„ë ˆì„ ìƒì„±
cs_df = pd.DataFrame(data = {'ì œëª©': board_titles})
```

## 13-4 ì •ê·œì‹ì„ í™œìš©í•œ í…ìŠ¤íŠ¸ ì •ì œ

```python
import re

## í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜
def clean_text(text):
    if isinstance(text, str):
       # íŠ¹ìˆ˜ ë¬¸ì ì œê±° regex (ì˜ì–´ ì†Œë¬¸ì, ì˜ì–´ ëŒ€ë¬¸ì, ìˆ«ì, í•œê¸€, ê³µë°±ê¸€ìë§Œ í—ˆìš©)
       text = re.sub(r'[^a-zA-Z0-9ê°€-í£\s]', '', text)
       # HTML íƒœê·¸ ì œê±° regex
       text = re.sub(r'<[^>]+>', '', text)
       # ì†Œë¬¸ìë¡œ ë³€í™˜ python í•¨ìˆ˜
       text = text.lower()
       return text
    else:
        return ""

## ë°ì´í„°í”„ë ˆì„ ì •ì œ í•¨ìˆ˜
def clean_df(df):
  if not df.empty:
      df['ì •ì œëœ ì œëª©'] = df['ì œëª©'].apply(clean_text)

## cs_dfì— ì •ì œ í•¨ìˆ˜ ì ìš©
clean_df(cs_df)
cs_df
```

## 13-6 í˜•íƒœì†Œ ë¶„ì„ì„ í†µí•œ í’ˆì‚¬ ë¶„ë¦¬

```python
from kiwipiepy import Kiwi

## í˜•íƒœì†Œ ë¶„ì„ê¸° ì´ˆê¸°í™” ë° ì‚¬ìš©ì ì‚¬ì „ ì¶”ê°€
kiwi = Kiwi()
kiwi.add_user_word('ë°©ì†¡ëŒ€', 'NNP')
kiwi.add_user_word('ë°©í†µëŒ€', 'NNP')

## í…ìŠ¤íŠ¸ í˜•íƒœì†Œ ë¶„ì„ í•¨ìˆ˜
def analyze_morphemes(text):
    if isinstance(text, str):
       result = kiwi.analyze(text)

       morphemes = []
       for token in result[0][0]:
            ## í˜•íƒœì†Œ(form)ì™€ í’ˆì‚¬(tag) ì •ë³´ë¥¼ íŠœí”Œë¡œ morphemesì— ì¶”ê°€
            morphemes.append((token['form'], token['tag']))
       return morphemes
    else:
        return []

## ë°ì´í„°í”„ë ˆì„ì— í˜•íƒœì†Œ ë¶„ì„ ì ìš© í•¨ìˆ˜
def pos_df(df):
  if not df.empty:
      # `ì •ì œëœ ì œëª©` ì¹¼ëŸ¼ì˜ ê°’ì— analyze_morphemes í•¨ìˆ˜ë¥¼ ì ìš©í•˜ê³  ê²°ê³¼ë¥¼ `í–‰íƒœì†Œ ë¶„ì„ ê²°ê³¼` ì¹¼ëŸ¼ì— ì €ì¥
        df['í˜•íƒœì†Œ ë¶„ì„ ê²°ê³¼'] = df['ì •ì œëœ ì œëª©'].apply(analyze_morphemes)

## í˜•íƒœì†Œ ë¶„ì„ ì ìš©
pos_df(cs_df)
cs_df
```

## 13-7 ëª…ì‚¬ ì¶”ì¶œ

```python
## í˜•íƒœì†Œ ë¶„ì„ ê²°ê³¼ì—ì„œ ëª…ì‚¬ ì¶”ì¶œ
def extract_nouns(morphemes):
  # í˜•íƒœì†Œê°€ NNG (ì¼ë°˜ëª…ì‚¬), NNP (ê³ ìœ ëª…ì‚¬) ì¸ wordë¥¼ nounsì— ì €ì¥
    nouns = [word for word, pos in morphemes if pos in ['NNG', 'NNP']]
  return nouns

## ë°ì´í„°í”„ë ˆì„ì— ëª…ì‚¬ ì¶”ì¶œ ì ìš©
def noun_df(df):
  if not df.empty:
      df['ëª…ì‚¬'] = df['í˜•íƒœì†Œ ë¶„ì„ ê²°ê³¼'].apply(extract_nouns)

## ëª…ì‚¬ ì¶”ì¶œ ì ìš©
noun_df(cs_df)
cs_df
```

## 13-8 ë°ì´í„° ì „ì²˜ë¦¬ í†µí•© í•¨ìˆ˜ ì •ì˜

```python
## ë°ì´í„°í”„ë ˆì„ ì „ì²˜ë¦¬ í†µí•© í•¨ìˆ˜
def preprocessing_df(df):
  clean_df(df) ## í…ìŠ¤íŠ¸ ì •ì œ ì ìš©
  pos_df(df) ## í˜•íƒœì†Œ ë¶„ì„ ì ìš©
  noun_df(df) ## ëª…ì‚¬ ì¶”ì¶œ ì ìš©
```

```python
from vllm import LLM

## LLM ëª¨ë¸ ë¡œë“œ ë° ì„¤ì •
llm = LLM(
    model="LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct",
    tensor_parallel_size=1,
    dtype="half",
    gpu_memory_utilization=0.6,
    max_model_len=32000,
    trust_remote_code=True # ì¶”ê°€ëœ ë¶€ë¶„
)
```

## 13-13 LLM ìƒ˜í”Œë§ íŒŒë¼ë¯¸í„° ì„¤ì • ë° í”„ë¡¬í”„íŠ¸ ìƒì„±

```python
from vllm import SamplingParams ## SamplingParams í´ë˜ìŠ¤ ì„í¬íŠ¸

## ìƒ˜í”Œë§ íŒŒë¼ë¯¸í„° ì„¤ì •
sampling_params = SamplingParams(
    temperature=0.3, ## ìƒì„± í…ìŠ¤íŠ¸ì˜ ë‹¤ì–‘ì„± ì¡°ì ˆ
    top_p=1.0, ## top_p ëˆ„ì  í™•ë¥  ë‚´ì—ì„œ í† í° ìƒ˜í”Œë§
    max_tokens=512, ## ìƒì„±ë  ìµœëŒ€ í† í° ìˆ˜
    frequency_penalty=0.5 ## ìì£¼ ë‚˜íƒ€ë‚˜ëŠ” í† í°ì— ëŒ€í•œ íŒ¨ë„í‹°
)

def format_prompt(user_input: str) -> str:
    ## í”„ë¡¬í”„íŠ¸ í˜•ì‹í™” í•¨ìˆ˜
    messages = [
        # ì‹œìŠ¤í…œ ë©”ì‹œì§€ ì¶”ê°€
        {"role": "system", "content": "You are EXAONE model from LG AI Research, a helpful assistant."},
        # ì‚¬ìš©ì ë©”ì‹œì§€ ì¶”ê°€
        {"role": "user", "content": user_input}
    ]
    return messages

## í”„ë¡¬í”„íŠ¸ ìƒì„±
prompt = format_prompt("ëŒ€í•œë¯¼êµ­ì˜ ìˆ˜ë„ëŠ” ì–´ë””ì¸ê°€ìš”? ìˆ˜ë„ì— ì—¬í–‰í•˜ëŸ¬ê°„ë‹¤ë©´ ì–´ë–¤ ì¦ê¸¸ê±°ë¦¬ê°€ ìˆì„ê¹Œìš”?")

## LLMì„ ì‚¬ìš©í•˜ì—¬ í…ìŠ¤íŠ¸ ìƒì„±
# chat í•¨ìˆ˜ë¥¼ ì´ìš© (í”„ë¡¬í”„íŠ¸ì™€ ìƒ˜í”Œë§ íŒŒë¼ë¯¸í„°)
outputs = llm.chat([prompt], sampling_params)

## ìƒì„±ëœ í…ìŠ¤íŠ¸ ì¶œë ¥
print("\nìƒì„±ëœ í…ìŠ¤íŠ¸:", outputs[0].outputs[0].text)
```

## 13-14 LLM ê¸°ë°˜ ë¶„ë¥˜ í”„ë¡¬í”„íŠ¸ ë° í•¨ìˆ˜ ì •ì˜

```python
## ë¶„ë¥˜ í”„ë¡¬í”„íŠ¸ í˜•ì‹í™” í•¨ìˆ˜
def format_classifier_prompt(board_title, title) -> str:
    messages = [
        {"role": "system", "content": "You are EXAONE model from LG AI Research, a helpful assistant."},
        {"role": "user", "content": """ì£¼ì–´ì§„ ê¸€ì˜ ì œëª©ì„ ë¶„ë¥˜í•˜ëŠ” AI ëª¨ë¸ì…ë‹ˆë‹¤.
ë¶„ë¥˜ í´ë˜ìŠ¤ëŠ” í•™ì‚¬/ì „ê³µ, í•™ìƒí™œë™, ì™¸ë¶€ì •ë³´ë¡œ ë‚˜ë‰˜ì–´ì§‘ë‹ˆë‹¤.
- í•™ì‚¬/ì „ê³µ: êµì¬, ê³¼ì œ, ì‹œí—˜ ë“± í•™ì—… ë° ì „ê³µ ê´€ë ¨ ë‚´ìš©
- í•™ìƒí™œë™: ë™ì•„ë¦¬, ìŠ¤í„°ë””, ëª¨ì„ ë“± í•™ìƒë“¤ì˜ ìë°œì  í™œë™
- ì™¸ë¶€ì •ë³´: ì·¨ì—…, ê³µëª¨ì „, í–‰ì‚¬ ë“± ì™¸ë¶€ ì •ë³´"""},
        {"role": "user", "content": """ì˜ˆì‹œ)
ì „ê³µëª…: ì»´í“¨í„°í•™ê³¼, ì œëª©: í”„ë¡œê·¸ë˜ë° ê³¼ì œ ì§ˆë¬¸ -> í•™ì‚¬/ì „ê³µ
ì „ê³µëª…: ì»´í“¨í„°í•™ê³¼, ì œëª©: ì•Œê³ ë¦¬ì¦˜ ìŠ¤í„°ë”” ëª¨ì§‘ -> í•™ìƒí™œë™
ì „ê³µëª…: ì»´í“¨í„°í•™ê³¼, ì œëª©: IT ì·¨ì—… ë°•ëŒíšŒ -> ì™¸ë¶€ì •ë³´"""},
        {"role": "user", "content": f"ì „ê³µëª…: {board_title}\nì œëª©: {title}"}
    ]
    return messages

## ë¶„ë¥˜ ìˆ˜í–‰ í•¨ìˆ˜
def classify(major, title):
    prompt = format_classifier_prompt(major, title) ## ë¶„ë¥˜ í”„ë¡¬í”„íŠ¸ ìƒì„±
    outputs = llm.chat([prompt], sampling_params) ## LLMìœ¼ë¡œ ë¶„ë¥˜ ì‹¤í–‰
    return outputs[0].outputs[0].text ## ë¶„ë¥˜ ê²°ê³¼ í…ìŠ¤íŠ¸ ë°˜í™˜
```

## 13-15 ë¶„ë¥˜ ê²°ê³¼ íŒŒì‹± í•¨ìˆ˜ ì •ì˜

```python
def get_class(analysis_result):
    ## ë¶„ì„ ê²°ê³¼ì—ì„œ ë¶„ë¥˜ í´ë˜ìŠ¤ ì¶”ì¶œ
    classes = ['í•™ì‚¬/ì „ê³µ', 'í•™ìƒí™œë™', 'ì™¸ë¶€ì •ë³´', 'ê¸°íƒ€']
    classes_index = [analysis_result.find(cls) for cls in classes]

    min_index = float('inf')
    min_class = 'ê¸°íƒ€'

    for i in range(len(classes)):
        current_index = classes_index[i]
        if current_index != -1 and current_index < min_index:
            min_index = current_index
            min_class = classes[i]

    return min_class
```

```python
## ì»´í“¨í„°ê³¼í•™ê³¼ ì œëª© ë¶„ë¥˜ í—¬í¼ í•¨ìˆ˜ ì •ì˜
def classify_cs(title):
  return classify('ì»´í“¨í„°ê³¼í•™ê³¼', title)

## 'ì œëª©' ì»¬ëŸ¼ì— ë¶„ë¥˜ í•¨ìˆ˜ ì ìš©í•˜ì—¬ ë¶„ì„ ê²°ê³¼ ì €ì¥
cs_df['class_analysis'] = cs_df['ì œëª©'].apply(classify_cs)
## ë¶„ì„ ê²°ê³¼ì—ì„œ ìµœì¢… ë¶„ë¥˜ í´ë˜ìŠ¤ ì¶”ì¶œí•˜ì—¬ ì €ì¥
cs_df['class'] = cs_df['class_analysis'].apply(get_class)
```

## 13-17 ë¶„ë¥˜ ê²°ê³¼ ì‹œê°í™”

```python
def plot_class(df):
  ## ë¶„ë¥˜ í´ë˜ìŠ¤ë³„ ê°œìˆ˜ ê³„ì‚° ë° ìˆœì„œ ì¬ì •ë ¬
  class_counts = df['class'].value_counts().reindex(['í•™ì‚¬/ì „ê³µ', 'í•™ìƒí™œë™', 'ì™¸ë¶€ì •ë³´'])

  plt.figure(figsize=(8, 6)) ## ê·¸ë˜í”„ í¬ê¸° ì„¤ì •
  plt.pie(class_counts,
          labels=class_counts.index,
          autopct='%.1f%%', ## í¼ì„¼íŠ¸ í‘œì‹œ í˜•ì‹
          startangle=90, ## ì‹œì‘ ê°ë„ ì„¤ì •
          )

  plt.title('Class Distribution') ## ê·¸ë˜í”„ ì œëª© ì„¤ì •
  plt.axis('equal') ## ì›í˜• ë¹„ìœ¨ ìœ ì§€
  plt.show() ## ê·¸ë˜í”„ í‘œì‹œ

plot_class(cs_df)
```
